1 Read the data from the csv file and create a DataFrame from it.

```python
import pandas as pd

eq_data = pd.read_csv("data_files/earthquakes.csv",delimiter=",")

eq_data.head()
```

2 Remove the following columns from the DataFrame: *Magnitude Seismic Stations*, *Azimuthal Gap*, *Horizontal Distance* and *Horizontal Error*.

```python
eq_data.drop(eq_data.iloc[:, 11:15],inplace=True, axis=1)
eq_data.head()
```

3 Filter data from the DataFrame so that only rows matching the following criteria will be shown:
    - Source must be US
    - Magnitude value must be over 7.8
    - ID value must start with the string "USP0007"

```python
eq_data = eq_data[(eq_data["Source"] == "US") & (eq_data["Magnitude"] >= 7.8) & (eq_data["ID"].str.startswith("USP0007"))]
eq_data
eq_data.dtypes
```

4 Create a new column to the tenth column position called Grade which will have values from the following formula: ("Magnitude" * "Depth") / "Root Mean Square". Results in this column should be presented with three decimals.

```python
#eq_data.insert(11, column = "Grade", value = eq_data[(eq_data["Magnitude"])*(eq_data["Depth"])/(eq_data["Root Mean Square"])])
#eq_data["Grade"] = eq_data[(eq_data['Magnitude'])*(eq_data['Depth'])/(eq_data['Root Mean Square'])]
eq_data["Grade"] = eq_data['Magnitude']*eq_data['Depth']/eq_data['Root Mean Square']
eq_data
#cols = ['Magnitude', 'Depth', 'Root Mean Square']
#set(eq_data.columns).issuperset(cols)
```
